```python
import cv2
import imutils
import numpy as np
import pandas as pd  # For creating and analyzing the table

# Load the image in color
path_img = "data/processed/frames/frame_0000.png"
original_img = cv2.imread(path_img)
if original_img is None:
    print(f"Error: Could not open or find the image at '{path_img}'")
    exit()

gray_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2GRAY)
blur_img = cv2.GaussianBlur(gray_img, (5, 5), 0)

# Apply edge detection
edged = cv2.Canny(blur_img, 50, 150)

# Dilate the edges (optional)
kernel = np.ones((3, 3), np.uint8)
dilated = cv2.dilate(edged, kernel, iterations=1)
```

```python
# Find the contours
contours = cv2.findContours(dilated.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
contours = imutils.grab_contours(contours)

# List to store contour properties
contour_properties = []

# Loop over the contours
for i, contour in enumerate(contours):
    # Calculate basic properties
    area = cv2.contourArea(contour)
    perimeter = cv2.arcLength(contour, True)
    x, y, w, h = cv2.boundingRect(contour)
    aspect_ratio = float(w) / h if h > 0 else float('inf')
    if aspect_ratio < 1:
        aspect_ratio = 1 / aspect_ratio

    # Calculate moments (for centroid, orientation, etc.)
    moments = cv2.moments(contour)
    centroid_x = int(moments['m10'] / moments['m00']) if moments['m00'] != 0 else -1
    centroid_y = int(moments['m01'] / moments['m00']) if moments['m00'] != 0 else -1

    # Calculate solidity
    solidity = float(area) / cv2.contourArea(cv2.convexHull(contour)) if cv2.contourArea(cv2.convexHull(contour)) > 0 else 0

    # Calculate extent
    extent = float(area) / (w * h) if (w * h) > 0 else 0

    # You can add more properties here, like Hu Moments for shape description

    properties = {
        "contour_index": i,
        "area": area,
        "perimeter": perimeter,
        "bounding_rect_x": x,
        "bounding_rect_y": y,
        "bounding_rect_width": w,
        "bounding_rect_height": h,
        "aspect_ratio": aspect_ratio,
        "centroid_x": centroid_x,
        "centroid_y": centroid_y,
        "solidity": solidity,
        "extent": extent,
        # Add more properties as needed
    }
    contour_properties.append(properties)
```

```python
# Create a Pandas DataFrame for easier analysis
df_contours = pd.DataFrame(contour_properties)

# Now you can analyze the 'df_contours' DataFrame
# print(df_contours.head())

# Example of basic analysis:
print("\nSummary Statistics:")
# print(df_contours.describe())

# You can further filter and analyze based on your observations
# For example, to find contours with high aspect ratio and small area:
potential_runway_lines = df_contours[(df_contours['aspect_ratio'] > 5) & (df_contours['area'] < 500)]
print("\nPotential Runway Lines (based on example criteria):")
print(potential_runway_lines)

# You can now iterate through the original contours based on the indices
# in your filtered DataFrame to draw them on the image
for index in potential_runway_lines['contour_index']:
    cv2.drawContours(original_img, [contours[index]], -1, (0, 255, 0), 2)

# Display the image with potential runway lines
cv2.imshow("Potential Runway Lines", original_img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

# Load the image in color
path_img = "data/processed/frames/frame_0000.png"
original_img = cv2.imread(path_img)
if original_img is None:
    print(f"Error: Could not open or find the image at '{path_img}'")
    exit()

gray_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2GRAY)
blur_img = cv2.GaussianBlur(gray_img, (5, 5), 0)
edged = cv2.Canny(blur_img, 50, 150)
kernel = np.ones((3, 3), np.uint8)
dilated = cv2.dilate(edged, kernel, iterations=1)
contours = cv2.findContours(dilated.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
contours = imutils.grab_contours(contours)

# List to store contour properties
contour_properties = []

# Loop over the contours and extract properties
for i, contour in enumerate(contours):
    area = cv2.contourArea(contour)
    perimeter = cv2.arcLength(contour, True)
    x, y, w, h = cv2.boundingRect(contour)
    aspect_ratio = float(w) / h if h > 0 else float('inf')
    if aspect_ratio < 1:
        aspect_ratio = 1 / aspect_ratio
    moments = cv2.moments(contour)
    centroid_x = int(moments['m10'] / moments['m00']) if moments['m00'] != 0 else -1
    centroid_y = int(moments['m01'] / moments['m00']) if moments['m00'] != 0 else -1
    solidity = float(area) / cv2.contourArea(cv2.convexHull(contour)) if cv2.contourArea(cv2.convexHull(contour)) > 0 else 0
    extent = float(area) / (w * h) if (w * h) > 0 else 0

    properties = {
        "contour_index": i,
        "area": area,
        "perimeter": perimeter,
        "bounding_rect_x": x,
        "bounding_rect_y": y,
        "bounding_rect_width": w,
        "bounding_rect_height": h,
        "aspect_ratio": aspect_ratio,
        "centroid_x": centroid_x,
        "centroid_y": centroid_y,
        "solidity": solidity,
        "extent": extent,
    }
    contour_properties.append(properties)

df_contours = pd.DataFrame(contour_properties)

# --- Plotting the image with contours and labels ---
output_img = original_img.copy()

for index, row in df_contours.iterrows():
    contour_index = int(row['contour_index'])
    centroid_x = int(row['centroid_x'])
    centroid_y = int(row['centroid_y'])
    aspect_ratio = row['aspect_ratio']
    area = int(row['area'])

    # Get the contour corresponding to the index
    contour = contours[contour_index]

    # Draw the contour
    cv2.drawContours(output_img, [contour], -1, (0, 255, 0), 2)

    # Add a label with some property (e.g., index, aspect ratio, or area)
    label = f"Idx: {contour_index}"  # You can change this label
    # label = f"AR: {aspect_ratio:.2f}"
    # label = f"Area: {area}"

    cv2.putText(output_img, label, (centroid_x - 20, centroid_y - 10),
                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 1, cv2.LINE_AA)

# Display the image with labeled contours
cv2.imshow("Image with Labeled Contours", output_img)
cv2.imwrite("contours_labels.png",output_img )
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# --- Define your filter criteria based on df_contours properties ---
min_aspect_ratio = 3
max_area = 500
min_solidity = 0.3

filtered_contours_df = df_contours[
    (df_contours['aspect_ratio'] >= min_aspect_ratio) &
    (df_contours['area'] <= max_area) &
    (df_contours['solidity'] >= min_solidity)
]

print(f"Number of contours before filtering: {len(df_contours)}")
print(f"Number of contours after filtering: {len(filtered_contours_df)}")
print("\nFiltered Contours DataFrame:")
print(filtered_contours_df)

# --- Plotting the image with the filtered contours highlighted ---
output_img = original_img.copy()
filtered_indices = filtered_contours_df['contour_index'].tolist()

for index in filtered_indices:
    if index < len(contours):
        cv2.drawContours(output_img, [contours[index]], -1, (0, 255, 0), 2)

# Display the image with filtered contours
cv2.imshow("Image with Filtered Contours", output_img)
cv2.waitKey(0)
cv2.destroyAllWindows()

# --- Optional: Plotting individual filtered contours in separate subplots ---
if not filtered_contours_df.empty:
    num_filtered = len(filtered_contours_df)
    fig, axes = plt.subplots(1, num_filtered, figsize=(15, 5))  # Adjust figure size as needed

    if num_filtered == 1:
        ax = axes
    else:
        ax = axes.flatten()

    for i, index in enumerate(filtered_contours_df['contour_index']):
        if index < len(contours):
            mask = np.zeros_like(cv2.cvtColor(original_img, cv2.COLOR_BGR2GRAY)) # Use grayscale dimensions
            cv2.drawContours(mask, [contours[index]], -1, 255, cv2.FILLED)
            masked_img = cv2.bitwise_and(original_img, original_img, mask=mask)

            # Get bounding box to crop the contour
            x, y, w, h = cv2.boundingRect(contours[index])
            cropped_contour = masked_img[y:y+h, x:x+w]

            if num_filtered == 1:
                ax.imshow(cv2.cvtColor(cropped_contour, cv2.COLOR_BGR2RGB))
                ax.set_title(f"Contour {index}")
                ax.axis('off')
            else:
                ax[i].imshow(cv2.cvtColor(cropped_contour, cv2.COLOR_BGR2RGB))
                ax[i].set_title(f"Contour {index}")
                ax[i].axis('off')

    plt.tight_layout()
    plt.show()
else:
    print("No contours matched the filter criteria.")
```
